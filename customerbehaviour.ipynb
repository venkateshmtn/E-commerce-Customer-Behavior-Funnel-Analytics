```python
#Create a copy (best practice)
df_clean=df.copy()
```

```python
#Standardize column names
df_clean.columns=(df_clean.columns.str.strip().str.lower().str.replace(' ','_'))
```

```python
#Clean numeric columns
df_clean['age']=df_clean['age'].astype('Int64')
df_clean['avg_spend_category']=df_clean['avg_spend_category'].astype('int')
```

```python
#Standardize ALL text columns
text_cols = df_clean.select_dtypes(include='object').columns

df_clean[text_cols]=df_clean[text_cols].apply(lambda x:x.str.strip().str.lower())
```

```python
#Handle missing values
df_clean['city']=df_clean['city'].fillna('unknown')
df_clean['frequent_clothing_product']=df_clean['frequent_clothing_product'].fillna('not specified')
df_clean['ott_subscription']=df_clean['ott_subscription'].fillna('no subscription')
df_clean['book_type']=df_clean['book_type'].fillna('not specified')
```

```python
df_clean.info()
```

```python
df_clean.isnull().sum()
```

```python
#Fill with median
df_clean['age']=df_clean['age'].fillna(df_clean['age'].median())
```

```python
df_clean['frequent_electronics_product'] = ( df_clean['frequent_electronics_product'].fillna('not applicable'))
df_clean['frequent_fashion_product']=(df_clean['frequent_fashion_product'].fillna('not applicable'))
```
